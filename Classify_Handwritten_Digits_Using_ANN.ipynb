# Classify Handwritten Digits Using ANN

```
# Importing Packages
import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# Importing Dataset
from keras.datasets import mnist
(train_image, train_labels), (test_image, test_labels) = mnist.load_data()

#Normalizing Dataset
train_image = keras.utils.normalize(train_image, axis=1)
test_image = keras.utils.normalize(test_image, axis =1)

# Building Model
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(300, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(10, activation="softmax"))

# Compiling Model
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=['accuracy'])

# Fitting the Model
model.fit(train_image, train_labels, epochs=10)

# Evaluate the Model
print(model.evaluate(test_image, test_labels))

# Predicting First 10 test images to check the model accuracy
pred = model.predict(test_image[:10])
#print(pred)
p=np.argmax(pred, axis=1)
print(p)
print(test_labels[:10])

# Visualizing prediction
for i in range(10):
  plt.imshow(test_image[i], cmap='binary')
  plt.title("Original: {}, Predicted: {}".format(test_labels[i], p[i]))
  plt.axis('off')
  plt.show()

```
